```python
import networkx as nx
import matplotlib.pyplot as plt
coord = {'Manager': [8, 0], 'Department Y head': [9, -1], 'Department X head': [7, -1], 'Staff A': [6, -2],'Staff B': [7, -2], 'Staff C': [8, -2],'Staff D': [9, -2], 'Staff E': [10, -2], 8: [13, -2]}
eg = [['Manager', 'Department X head'], ['Manager','Department Y head'], ['Department X head', 'Staff A'],['Department X head', 'Staff B'], ['Department Y head','Staff C'], ['Department Y head','Staff D'], ['Department Y head', 'Staff E']]
G=nx.DiGraph(name='test me')
G.add_edges_from(eg) 
colors=["skyblue","lightgreen","lightgreen","orange","orange","orange","orange","orange"]
nx.draw(G, with_labels=True, node_size=2000, node_color=colors, pos=coord) 
plt.show()

ne=G.number_of_edges()
print('number of edges=',ne)
nn=G.number_of_nodes()
print('number of nodes=',nn)

first, *middle, last = nx.topological_generations(G)
print(f'Start node:  {first}')
print(f'End nodes: {last}')
paths1 = list(nx.all_simple_paths(G, 'Manager', 'Staff A'))
paths2 = list(nx.all_simple_paths(G, 'Manager', 'Staff B'))
paths3 = list(nx.all_simple_paths(G, 'Manager', 'Staff C'))
paths4 = list(nx.all_simple_paths(G, 'Manager', 'Staff D'))
paths5 = list(nx.all_simple_paths(G, 'Manager', 'Staff E'))
print('-------------------------------------------------')
print('list of all paths from the root to the leaves of the tree.')

print(f'Simple paths from root to Staff A: {paths1}')
print(f'Simple paths from root to Staff B: {paths2}')
print(f'Simple paths from root to Staff C: {paths3}')
print(f'Simple paths from root to Staff D: {paths4}')
print(f'Simple paths from root to Staff E: {paths5}')
```
